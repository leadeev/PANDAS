- Which actor is mentioned the most often?


```

import pandas as pd

link = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv'
df = pd.read_csv(link)
df
```

```

df['actors_list']
```

```

df['actors_list'].iloc[0]     
```

```
actors = df['actors_list'].iloc[0]
eval(actors)
```

```

def actors(actor):
  return eval(actor)
```

```

df['actors_list'] = df['actors_list'].apply(actors)
```

```

df
```

```

df['actors_list'].explode()
```

```

df_actors = df.explode('actors_list')
df_actors
```

```

df_actors['actors_list'].value_counts()
```

```

df_actors['actors_list'].mode()
```
